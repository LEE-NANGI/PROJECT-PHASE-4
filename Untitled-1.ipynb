The function below loads all the datasets and merges them so that we could later clean them. 

```python
# loading all the dataset and merging them. 

def merge_datasets(file_paths, merge_column):

    # Check if the file_paths list is empty
    if not file_paths:
        raise ValueError("The file_paths list is empty.")

    # Initialize an empty DataFrame for the merged dataset
    merged_dataframe = pd.DataFrame()

    # Iterate through the file paths and merge datasets
    for file_path in file_paths:
        # Read each dataset into a DataFrame
        dataset = pd.read_csv(file_path)

        # Merge the dataset with the existing merged_dataframe
        if merged_dataframe.empty:
            merged_dataframe = dataset
        else:
            merged_dataframe = pd.merge(merged_dataframe, dataset, on=merge_column, how='inner')

    return merged_dataframe

# List of file paths to the datasets
dataset_paths = ['ml-latest-small/movies.csv', "ml-latest-small/ratings.csv",]  #'ml-latest-small/tags.csv', 'ml-latest-small/links.csv']

# The column on which you want to merge the datasets on
merge_column = "movieId"

# Call the function to merge the datasets
merged_data = merge_datasets(dataset_paths, merge_column)

# The merged dataset
merged_data.head()
```

## Data Cleaning

The data provided has 100836 rows and 5 columns, I did not merge the other two datasets(`links.csv` and `tags.csv`) as they added more rows to our dataset that were not necessary for our analysis. Therefore let us proceed to cleaning. 

```python
# This  function checks for missing values and removes them and shows a dataframe for their percentages

def missing_data(data):
    
    # finding sum of all the missing data in the df.
    miss = data.isnull().sum().sort_values(ascending=False)
    
    # Converting it into percentage. 
    miss_percentage = miss / len(data)
    
    # Creating a dataframe to show the missing data. 
    df_missing = pd.DataFrame({'Missing values number': miss, 'Percentage missing values': miss_percentage})
    
    # drop missing values
    df_missing.drop(df_missing[df_missing['Percentage missing values'] == 0].index, inplace=True)
    
    return df_missing

missing = missing_data(merged_data)
missing
```

```python
 # Function to check for duplicates.

def duplicates(data):   
    duplicates = data.duplicated().sum()
    return duplicates

df_dup = duplicates(merged_data)
df_dup
```

## Modelling

In this section, we transition from data cleaning and preparation to the pivotal step of constructing predictive models. For our project on recommendation system, this phase is crucial as it allows us to develop accurate models that can recommend 5 best movies to a user. 

During this stage, I will utilize the cleaned and prepared "data" dataset, ensuring it is well-suited for classification analysis. I will consider only a few columns for recommendation.

To commence, we are going to pick the relevant columns for modelling.

```python
data = df[['userId', 'movieId', 'rating']]
```

```python
# Create a Surprise Reader object
reader = Reader(rating_scale=(0.5, 5.0))

# Load the dataset from the DataFrame
data = Dataset.load_from_df(data, reader)
```

```python
# Split data into training and testing sets
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)
```

#### KNNBasic model

KNNBasic is a simpler variant of collaborative filtering based on the basic K-Nearest Neighbors algorithm. It focuses solely on finding the nearest neighbors without considering baseline estimates.

```python
knn = KNNBasic(random_state=42)
knn.fit(trainset)
```

```python
predictions = knn.test(testset)
print(accuracy.rmse(predictions))
print(accuracy.mse(predictions))
```

```python
# Using cross validation and gridsearchCV to improve the above model

param_grid = {
    'k': [20, 30, 40],  # Number of neighbors
    'sim_options': {
        'name': ['cosine', 'msd', 'pearson'],
        # 'min_support': [1, 5],  # Minimum number of common items for similarity
        'user_based': [True, False]  # User-based or item-based
    }
}

gs_basic = GridSearchCV(KNNBasic, param_grid, measures=['rmse', 'mae'], cv=3)
gs_basic.fit(data)
```

```python
# Get the best parameters and their corresponding RMSE score
best_params = gs_basic.best_params['rmse']
best_rmse = gs_basic.best_score['rmse']

# Print the best parameters and RMSE score
print("Best Parameters:", best_params)
print("Best RMSE Score:", best_rmse)
```

```python
# Now, you can use the 'trainset' and 'testset' for evaluation

KNNbasic_model = gs_basic.best_estimator['rmse']
KNNbasic_model.fit(trainset)
predictions = KNNbasic_model.test(testset)
KNNbasic_rmse = accuracy.rmse(predictions)
KNNNbasic_mae = accuracy.mae(predictions)
print("RMSE on Test Set:", KNNbasic_rmse)
print("MAE on Test Set:", KNNNbasic_mae)
```

#### SVD

__Singular-Value Decomposition__ or SVD is a common and widely used matrix decomposition method. It is a matrix factorization technique that decomposes the user-item interaction matrix into three matrices, which represent latent factors for users and items. These latent factors capture hidden patterns in the data. All matrices are able to be factored using an SVD, which makes it more stable than other methods, such as the eigendecomposition. As such, it is often used in a wide array of applications including compression and data reduction.

In simple terms, SVD is the factorization of a matrix into 3 matrices. So if we have a matrix A, then its SVD is represented by the equation:

$$ A = U\Sigma V^T$$

```python
svd = SVD(random_state=42)
svd.fit(trainset)
```

```python
predictions = svd.test(testset)
print(accuracy.rmse(predictions))
print(accuracy.mse(predictions))
```

```python
param_grid = {'n_factors':[20, 100],
              'n_epochs': [5, 10], 
              'lr_all': [0.002, 0.005],
              'reg_all': [0.4, 0.6]}

gs_svd_model = GridSearchCV(SVD, param_grid=param_grid, n_jobs = -1, joblib_verbose=5)
gs_svd_model.fit(data)
```

```python
# Get the best parameters and their corresponding RMSE score
best_params = gs_svd_model.best_params['rmse']
best_rmse = gs_svd_model.best_score['rmse']

# Print the best parameters and RMSE score
print("Best Parameters:", best_params)
print("Best RMSE Score:", best_rmse)
```

```python
# Now, you can use the 'trainset' and 'testset' for evaluation
svd_model = gs_svd_model.best_estimator['rmse']
svd_model.fit(trainset)
predictions = svd_model.test(testset)
SVD_rmse = accuracy.rmse(predictions)
SVD_mae = accuracy.mae(predictions)
print("RMSE on Test Set:", SVD_rmse)
print("MAE on Test Set:", SVD_mae)
```

SVD performs slightly better than KNNBasic in terms of RMSE and MAE. It offers a reasonable level of accuracy in predicting user preferences. While it's an improvement over KNNBasic, there may still be room for further enhancement.

#### KNNBaseline

 KNNBaseline is a collaborative filtering algorithm that leverages the idea of finding similar users or items to make recommendations. It extends the basic K-Nearest Neighbors approach by incorporating a baseline estimate, which considers the overall trends in user-item interactions.

```python
knn_baseline = KNNBaseline(random_state=42)
knn_baseline.fit(trainset)
```

```python
predictions = knn_baseline.test(testset)
print(accuracy.rmse(predictions))
print(accuracy.mse(predictions))
```

```python
param_grid = {
    'k': [20, 30, 40],  # Number of neighbors
    'sim_options': {
        'name': ['cosine', 'msd', 'pearson'],
        'min_support': [1, 5],  # Minimum number of common items for similarity
        'user_based': [True, False]  # User-based or item-based
    },
    'bsl_options': {
        'method': ['als', 'sgd'],  # Baseline method (Alternating Least Squares or Stochastic Gradient Descent)
        'reg': [0.02, 0.1, 0.2]  # Regularization term
    }
}


gs_base = GridSearchCV(KNNBaseline, param_grid, measures=['rmse', 'mae'], cv=3)
gs_base.fit(data)
```

```python
# Get the best parameters and their corresponding RMSE score
best_params = gs_base.best_params['rmse']
best_rmse = gs_base.best_score['rmse']

# Print the best parameters and RMSE score
print("Best Parameters:", best_params)
print("Best RMSE Score:", best_rmse)
```

```python
# Now, you can use the 'trainset' and 'testset' for evaluation
KNNbaseline_model = gs_base.best_estimator['rmse']
KNNbaseline_model.fit(trainset)
predictions = KNNbaseline_model.test(testset)
KNNbaseline_rmse = accuracy.rmse(predictions)
KNNbaseline_mae = accuracy.mae(predictions)
print("RMSE on Test Set:", KNNbaseline_rmse)
print("MAE on Test Set:", KNNbaseline_rmse)
```

KNNBaseline outperforms both KNNBasic and SVD in terms of RMSE and MAE, making it the most accurate model among the three. However, it's important to note that it took a significantly longer time (300 mins) to compute, which might not be practical for real-time recommendations or scenarios where computational efficiency is crucial.

Using the best model, here is a few lines of code to be able to see the best recommendations from given the `userId`.

```python
# Select the best model based on your evaluation results 
best_model = KNNBaseline()

# Train the best model on the entire dataset
best_model.fit(data.build_full_trainset())

# Define a user for whom you want to make recommendations 
user_id = 1000

# Get the movies the user has already rated
user_rated_movies = df[df['userId'] == user_id]['movieId'].tolist()

# Create a list of movies the user hasn't rated (the anti-test set)
anti_testset = []
for movie_id in df['movieId'].unique():
    if movie_id not in user_rated_movies:
        anti_testset.append((user_id, movie_id, 0.0))  # Set the initial rating to 0.0

# Predict ratings for the unrated movies
predictions = best_model.test(anti_testset)

# Sort the predictions by predicted rating in descending order
predictions.sort(key=lambda x: x.est, reverse=True)

# Get the top 5 movie recommendations
top_recommendations = predictions[:5]

# Print the top 5 movie recommendations
for prediction in top_recommendations:
    movie_id = prediction.iid
    predicted_rating = prediction.est
    movie_title = df[df['movieId'] == movie_id]['Title'].values[0]
    print(f"Movie: {movie_title} (Movie ID: {movie_id}), Predicted Rating: {predicted_rating}")
```

The code above gets User Ratings For a specific user, then gathers their ratings for the movies they've already watched and rated. These ratings will serve as the basis for generating recommendations.

It generates Movie Recommendations using the best trained model to generate movie recommendations for the user based on their historical ratings. Typically, you can use the `anti_testset` method to create a list of movies the user hasn't rated, and then predict ratings for these unrated movies. Finally, it selects the top 5 movies with the highest predicted ratings as recommendations.
